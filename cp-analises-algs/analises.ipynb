### Importando e definindo funções para o projeto

```python
from algs import *

def pega_temp_alg(alg, lista, *args):
    '''
    Função pega_temp_alg recebe um algoritmo, uma lista e os argumentos necessários para o algoritmo
    e retorna uma lista dos tempos de execução dos algoritmos ao ordenar as listas.
    
    parametros: 
        alg - algoritmo a ser testado
        lista - lista a ser ordenada
        *args - argumentos necessários para o algoritmo
    retorno:
        tempos - lista de tempos de execução do algoritmo
    '''
    tempos = []
    for i in range(20):
        tempos.append(get_time(alg, lista, *args))
    return tempos

def calc_media(tempos):
    '''
    Função calc_media recebe uma lista de tempos e retorna a média dos tempos.
    parametros:
        tempos - lista de tempos
    retorno: 
        media - média dos tempos
    '''
    media = sum(tempos) / len(tempos)
    return media
```

```python
def pega_lista(nm_arq):
    '''
    Função pega_lista recebe o nome de um arquivo e retorna uma lista com os valores lidos do arquivo.
    parametro: 
        nm_arq - nome do arquivo
    retorno:
        lista - lista com os valores lidos do arquivo
    '''
    lista =le_arquivo(nm_arq)
    return lista

def cria_dfs(alg, lista, label, *args):
    '''
    Função cria_dfs recebe um algoritmo, uma lista, um label e os argumentos necessários para o algoritmo e 
    cria um dataframe com os tempos de execução do algoritmo e outro com a média dos tempos.
    parametros:
        alg - algoritmo a ser testado
        lista - lista a ser ordenada
        label - label do dataframe
        *args - argumentos necessários para o algoritmo
    retorno:
        df_tempos - dataframe com os tempos de execução do algoritmo
        df_media - dataframe com a média dos tempos
    '''
    df_tempos = pd.DataFrame(data={label: pega_temp_alg(alg, lista, *args)})
    df_medias = pd.DataFrame(data={label: calc_media(df_tempos[label])}, index=[0])
    return df_tempos, df_medias

```

## Bubble Sort
Bubble Sort é um algoritmo de utilizado para ordenar listas, e para isso ele percorre a lista comparando o elemento atual com o próximo, se o elemento atual for maior que o próximo, então os elementos são trocados de posição. O algoritmo continua percorrendo a lista até que não haja mais trocas de elementos, o que significa que a lista está ordenada.

Criando os dataframes

```python

# lista com mil valores
df_tempos_1, df_medias_1 = cria_dfs(bubble_sort, pega_lista('lista1.txt'),'mil')

# lista com cinco mil valores
df_tempos_5, df_medias_5 = cria_dfs(bubble_sort, pega_lista('lista5.txt'), '5 mil')

# lista com dez mil valores
df_tempos_10, df_medias_10 = cria_dfs(bubble_sort, pega_lista('lista10.txt'), '10 mil')

# lista com cinquenta mil valores
df_tempos_50, df_medias_50 = cria_dfs(bubble_sort, pega_lista('lista50.txt'), '50 mil')

# lista com cem mil valores
df_tempos_100, df_medias_100 = cria_dfs(bubble_sort, pega_lista('lista100.txt'), '100 mil')

# contcatena os dataframes
df_tempos_bubble = pd.concat([df_tempos_1, df_tempos_5, df_tempos_10, df_tempos_50, df_tempos_100], axis=1)
df_tempos_bubble['Sort'] = 'Bubble Sort'
df_medias_bubble = pd.concat([df_medias_1, df_medias_5, df_medias_10, df_medias_50, df_medias_100], axis=1)
df_tempos_bubble['Sort'] = 'Bubble Sort'

```

Persistindo dados em arquivos csv

```python
df_tempos_bubble.to_csv('tempos_bubble.csv', index=False)
df_medias_bubble.to_csv('medias_bubble.csv', index=False)
```

## Insertion Sort
Este algoritmo funciona de maneira muito similar quando ordenamos as cartas em nossas mãos quando estamos jogando. Ele percorre a lista e a "separa" em uma parte ordenada e outra desordenada, a cada iteração ele pega o primeiro elemento da parte desordenada e o coloca na parte ordenada, de forma que a parte ordenada fique sempre ordenada. O algoritmo continua percorrendo a lista até que não haja mais trocas de elementos, o que significa que a lista está ordenada.

## Merge Sort
Merge Sort é um algoritmo de ordenação que utiliza a estratégia de dividir para conquistar. Ele divide a lista em duas partes, ordena cada uma das partes e depois as mescla. Para isso, o algoritmo utiliza da recursividade para divir as partes da lista até que cada parte tenha apenas um elemento, e então ele começa a mesclar as partes ordenadas. O algoritmo continua percorrendo a lista até que não haja mais trocas de elementos, o que significa que a lista está ordenada. 

```python
# lista com mil valores
df_tempos_1, df_medias_1 = cria_dfs(insertion_sort, pega_lista('lista1.txt'),'mil')

# lista com cinco mil valores
df_tempos_5, df_medias_5 = cria_dfs(insertion_sort, pega_lista('lista5.txt'), '5 mil')

# lista com dez mil valores
df_tempos_10, df_medias_10 = cria_dfs(insertion_sort, pega_lista('lista10.txt'), '10 mil')

# lista com cinquenta mil valores
df_tempos_50, df_medias_50 = cria_dfs(insertion_sort, pega_lista('lista50.txt'), '50 mil')

# lista com cem mil valores
df_tempos_100, df_medias_100 = cria_dfs(insertion_sort, pega_lista('lista100.txt'), '100 mil')

# contcatena os dataframes
df_tempos_merge = pd.concat([df_tempos_1, df_tempos_5, df_tempos_10, df_tempos_50, df_tempos_100], axis=1)
df_tempos_merge['Sort'] = 'Merge Sort'
df_medias_merge = pd.concat([df_medias_1, df_medias_5, df_medias_10, df_medias_50, df_medias_100], axis=1)
df_tempos_merge['Sort'] = 'Merge Sort'
```

Persistindo dados em arquivos csv

```python
df_tempos_merge.to_csv('tempos_merge.csv', index=False)
df_medias_merge.to_csv('medias_merge.csv', index=False)
```

## Quick Sort
O Quick Sort, assim como o Merge Sort, também utiliza a estratégia de dividir para conquistar. Ele divide a lista em duas partes: à esquerda os valores menores que o pivô e à direita os maiores que o pivô. Para isso, o algoritmo utiliza da recursividade para divir as partes da lista até uma parte tenha apenas um elemento, e então ele começa a mesclar as partes ordenadas. 


```python
# lista com mil valores
df_tempos_1, df_medias_1 = cria_dfs(quick_sort, pega_lista('lista1.txt'), 'mil', 0, len(pega_lista('lista1.txt'))-1)

# lista com cinco mil valores
df_tempos_5, df_medias_5 = cria_dfs(quick_sort, pega_lista('lista5.txt'), '5 mil', 0, len(pega_lista('lista5.txt'))-1)

# lista com dez mil valores
df_tempos_10, df_medias_10 = cria_dfs(quick_sort, pega_lista('lista10.txt'), '10 mil', 0, len(pega_lista('lista10.txt'))-1)

# lista com cinquenta mil valores
df_tempos_50, df_medias_50 = cria_dfs(quick_sort, pega_lista('lista50.txt'), '50 mil', 0, len(pega_lista('lista50.txt'))-1)

# lista com cem mil valores
df_tempos_100, df_medias_100 = cria_dfs(quick_sort, pega_lista('lista100.txt'), '100 mil', 0, len(pega_lista('lista100.txt'))-1)

# contcatena os dataframes
df_tempos_quick = pd.concat([df_tempos_1, df_tempos_5, df_tempos_10, df_tempos_50, df_tempos_100], axis=1)
df_tempos_quick['Sort'] = 'Quick Sort'
df_medias_quick = pd.concat([df_medias_1, df_medias_5, df_medias_10, df_medias_50, df_medias_100], axis=1)
df_tempos_quick['Sort'] = 'Quick Sort'
```
